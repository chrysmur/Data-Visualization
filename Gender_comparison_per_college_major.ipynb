```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
fig = plt.figure(figsize=(16,20))
#display all the category comparison
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
all_cats = [stem_cats,other_cats,lib_arts_cats]
counter =[]
for major_cat in all_cats:
    for i, cat in enumerate(major_cat):
        counter.append(i)
        counter_len = len(counter)
        ax = "ax%s"%(counter_len)
        ax = fig.add_subplot(6,3,counter_len)
        ax.plot(women_degrees['Year'], women_degrees[cat], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[cat], c=cb_orange, label='Men', linewidth=3)
        ax.tick_params(bottom="off", top="off", left="off", right="off")
        #show only 0 and 100 in the y axis and add center line
        ax.set_yticks([0,100])
        ax.axhline(50, c = (171/255,171/255,171/255), alpha = 0.3)
        #remove x axis except for the bottom plots
        if counter_len<15:
            ax.tick_params(labelbottom = 'off')
        ax.set_title(cat)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        for key,spine in ax.spines.items():
            spine.set_visible(False)
            
            #dd  3 annotations
        if counter_len == 1:
            ax.text(2005, 85, 'Women')
            ax.text(2005, 30, 'Men')
        elif counter_len == 6:
            ax.text(2005, 85, 'Men')
            ax.text(2002, 20, 'Women')
        elif counter_len == 15:
            ax.text(2005, 40, 'Men')
            ax.text(2002, 70, 'Women')            

            plt.savefig('comparison.png')
plt.show()
    
            
    
    
```
